```python
import matplotlib.pyplot as plt
import numpy as np

# Data jumlah pembeli harian (daily number of customers) untuk bulan Desember 2019
dates = np.arange(1, 32)
customers = np.random.randint(50, 200, size=31)  # Menghasilkan data acak antara 50 dan 200 untuk contoh

# Membuat ukuran figure 10x5
plt.figure(figsize=(10, 5))

# Membuat line chart
plt.plot(dates, customers, marker='o', linestyle='-', color='b')

# Menambahkan title dan axis labels
plt.title('Jumlah Pembeli Harian - Desember 2019', fontsize=14, fontweight='bold')
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Jumlah Pembeli Unik', fontsize=12)

# Menambahkan grid untuk memperjelas pembacaan
plt.grid(True)

# Menyesuaikan tampilan tick marks
plt.xticks(ticks=np.arange(1, 32, 2))
plt.yticks(ticks=np.arange(0, max(customers)+10, 20))

# Menampilkan chart
plt.show()
```
